## Remove the index column

```python
data.drop(['index'], axis= 1, inplace= True)
data.shape
```

## Splitting the Salary Estimate to 2 columns
#### -We are splitting the column to Min Salar and Max Salary
#### -Creating a Mean_salary column

```python
data['Salary Estimate'] = data['Salary Estimate'].str.replace(' (Glassdoor est.)','')
data['Salary Estimate'].head()
```

```python
data[['Min_Salary', 'Max_Salary']] = data['Salary Estimate'].str.split('-', expand=True)
data[['Min_Salary', 'Max_Salary']].head()
```

#### As we can see from the cell below that there are other string values in the Max_Salary column that need to be removed before trimming and changing the dtype

```python
print(data['Min_Salary'].unique())
print(data['Max_Salary'].unique())
```

```python
data['Max_Salary'] = data['Max_Salary'].str.replace('(Employer est.)','')
```

```python
data[['Min_Salary', 'Max_Salary']] = data[['Min_Salary', 'Max_Salary']].apply(lambda x: x.str.strip('$K')) 
data[['Min_Salary', 'Max_Salary']] = data[['Min_Salary', 'Max_Salary']].astype(int)
data[['Min_Salary', 'Max_Salary']] = data[['Min_Salary', 'Max_Salary']] * 1000
print(data['Min_Salary'].dtype, data['Max_Salary'].dtype)
```

```python
data[['Min_Salary', 'Max_Salary']].head()
```

```python
data['Average_Salary'] = (data['Min_Salary'] + data['Max_Salary']) / 2
```

```python
data.drop(columns= ['Salary Estimate'], axis=1, inplace= True)
```

##  Size of employees:
#### - change to Mean of employees
#### - Change the dtype to int 

```python
data['Size'] = data['Size'].str.replace(' employees','')
```

```python
data['Size'].unique()
```

```python
data_map = {'1001 to 5000': 2500, '5001 to 10000': 5000, '501 to 1000': 750, '51 to 200': 125,
       '10000+': 10000, '201 to 500': 350, '1 to 50': 25, '-1': 0, 'Unknown': 0}
data['Size'] = data['Size'].map(data_map)
data.rename(columns={'Size':'Average_size_of_employees'}, inplace= True)
data['Average_size_of_employees'].head()
```

#### We have some companies with 0 number of employees which does not make sense

```python
data['Average_size_of_employees'].value_counts()
```

#### List of companies with no employee range

#### As we can see there are many company names are redundunt so we need to drop the duplicate companies and remove the companies with no names 

```python
print(data.shape)
data.drop_duplicates(subset=['Company Name'], keep= 'first', inplace=True)
print(data.shape)
data = data.drop(data[(data['Company Name'] == '') | (data['Company Name'] == ' ')].index)
print(data.shape)
```

### The companies with no number of employees might be classified data or startups. Whether we impute or drop those companies rely on the domain knowledge and the specific analysis we might do later.

## Revnue Column
#### The Revenue column has to a number so we can do analysis on different comapnies in the future. We changed every value in the Revenue column to represent roughly the average value of the companies revenue.

```python
data['Revenue'].unique()
```

```python
revenue_map = {'Unknown / Non-Applicable':0, '$1 to $2 billion (USD)':1.5*(10**3),
       '$100 to $500 million (USD)': 300, '$10+ billion (USD)': 10**4,
       '$2 to $5 billion (USD)': 3.5*(10**3), '$500 million to $1 billion (USD)': 750,
       '$5 to $10 billion (USD)': 7.5*(10**3), '$10 to $25 million (USD)': 17.5,
       '$25 to $50 million (USD)': 37.5, '$50 to $100 million (USD)': 75,
       '$1 to $5 million (USD)': 3, '$5 to $10 million (USD)': 7.5,
       'Less than $1 million (USD)': 1, '-1':0}
data['Revenue_in_Million'] = data['Revenue'].map(revenue_map)
data['Revenue_in_Million'].dtype
```

```python
data.drop(columns=['Revenue'], axis= 1, inplace= True)
```

### In the cell below there are 155 companies with 0 revenue which indicates that it is either classified information or startup companies. One way to do this is to impute those missing values using the mean of the revenue.
### You might also leave it as is if it is not that important in your analysis.

```python
data['Revenue_in_Million'].value_counts()
```

```python
mask = data['Revenue_in_Million'] == 0
data.loc[mask, 'Revenue_in_Million'] = data['Revenue_in_Million'].mean()
```

```python
data['Revenue_in_Million'].value_counts()
```

## Exracting important keywords from the Description column
#### The Description columns contains information about the job responsibilities and the technologies used in the applied position.

```python
print(data['Job Description'][0])
print(data['Job Description'][1])
print(data['Job Description'][2])
```

```python
data['Python'] = data['Job Description'].apply(lambda x: 1 if 'python' in x.lower() else 0)
data['Excel'] = data['Job Description'].apply(lambda x: 1 if 'excel' in x.lower() else 0)
data['AWS'] = data['Job Description'].apply(lambda x: 1 if 'aws' in x.lower() else 0)
data['Tableau'] = data['Job Description'].apply(lambda x: 1 if 'tableau' in x.lower() else 0)
data['SQL'] = data['Job Description'].apply(lambda x: 1 if 'sql' in x.lower() else 0)
data['Hadoop'] = data['Job Description'].apply(lambda x: 1 if 'hadoop' in x.lower() else 0)
```

## Age of the company from the founded column by subtracting it from the current year
#### Notice that some company's has -1 in the Founded column those we will set them to the current year

```python
data['Founded'].value_counts()[0:5]
```

```python
current_year = 2023
data['Founded'] = data['Founded'].apply(lambda x: current_year if x == -1 else x)
data['Company_age'] = current_year - data['Founded']
data['Company_age'].head()
```

## Headquarters and Location columns splitting the city from the state/province
#### We can see from the cell below that the Location column contains United States, Remote, New Jersey, Utah, and Texas as values.

```python
data['Location'].value_counts()
```

```python
data['Job_state'] = data['Location'].apply(lambda x: x.split(',')[-1])
data['Job_state'].head()
```

```python
data['Job_state'].value_counts()
```

```python
data[(data['Location'] == 'United States') | (data['Location'] == 'Remote') | (data['Location'] == 'New Jersey')\
     | (data['Location'] == 'Utah') | (data['Location'] == 'Texas') | (data['Location'] == 'California')][['Location', 'Headquarters']]
```

```python
def work_location_cleaning(title):
    if 'United States' == title:
        title = 'US'
    elif 'New Jersey' == title:
        title = 'NJ'
    elif 'Texas' == title:
        title = 'TX'
    elif 'Utah' == title:
        title = 'UT'
    elif 'California' == title:
        title = 'CA'
    return title
    
data['Job_state'] = data['Job_state'].map(work_location_cleaning) 
data['Job_state'].unique()
```

## Rating column

```python
data['Rating'].value_counts().head()
```

```python
data['Rating'] = data['Rating'].map(lambda x: 0 if x == -1 else x)
data['Rating'].value_counts().head()
```

## Now for the -1 values in the headquarter.
### We will fill those values with the value in the location column

```python
data['Headquarters'].value_counts().head()
```

```python
mask = data['Headquarters'] == '-1'
location_df = data.loc[mask, 'Location']
data.loc[mask, 'Headquarters'] = location_df
data['Headquarters'].value_counts().head()
```
